```python
zip_path = "/content/Youtube Channel Info.zip"
extract_path = "/content/"
os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'Channel Name': ['Ed Sheeran', 'Blackpink', 'MrBeast', 'Colors TV', 'SET India', 'Sony Music India', 'Wave Music', 'Zee TV', '5-Minute Crafts', 'Aaj Tak', 'Justin Bieber'],
    'Subscribers': [1000000, 2800000, 400000, 5700000, 4000000, 6200000, 7800000, 8800000, 9000000, 10000000, 11000000]
})

if 'Channel Name' in df.columns and 'Subscribers' in df.columns:
    top10 = df.nlargest(10, 'Subscribers').sort_values(by='Subscribers', ascending=True)
    top10['Subscribers_m'] = top10['Subscribers'] / 1_000_000

    fig, ax = plt.subplots(figsize=(10,6))
    sns.barplot(x='Subscribers_m', y='Channel Name', data=top10, palette='viridis', ax=ax)
    ax.set_title('Top 10 YouTube Channels by Subscribers', fontsize=16)
    ax.set_xlabel('Subscribers (Millions)', fontsize=12)
    ax.set_ylabel('Channel Name', fontsize=12)

    for i, value in enumerate(top10['Subscribers_m']):
        ax.text(value + 0.1, i, f'{top10["Subscribers"].iloc[i]:,}', va='center', fontsize=10)

    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'Channel Name': ['Ed Sheeran', 'Blackpink', 'MrBeast', 'Colors TV', 'SET India', 'Sony Music India', 'Wave Music', 'Zee TV', '5-Minute Crafts', 'Aaj Tak', 'Justin Bieber'],
    'Subscribers': [1000000, 2800000, 400000, 5700000, 4000000, 6200000, 7800000, 8800000, 9000000, 10000000, 11000000]
})

if 'Channel Name' in df.columns and 'Subscribers' in df.columns:
    top10 = df.nlargest(10, 'Subscribers').sort_values(by='Subscribers', ascending=True)
    top10['Subscribers_m'] = top10['Subscribers'] / 1_000_000

    plt.figure(figsize=(10,6))
    sns.histplot(data=top10, x='Subscribers_m', bins=10, kde=True, color='purple')

    plt.title('Top 10 YouTube Channels by Subscribers', fontsize=16)
    plt.xlabel('Subscribers (Millions)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({
    'Channel Name': ['Ed Sheeran', 'Blackpink', 'MrBeast', 'Colors TV', 'SET India', 'Sony Music India',
                     'Wave Music', 'Zee TV', '5-Minute Crafts', 'Aaj Tak', 'Justin Bieber'],
    'Subscribers': [1000000, 2800000, 400000, 5700000, 4000000, 6200000, 7800000, 8800000, 9000000, 10000000, 11000000],
    'View Count': [150000000, 800000000, 450000000, 700000000, 950000000, 880000000,
                   600000000, 400000000, 920000000, 980000000, 1100000000]  # Example view counts
})


if 'Channel Name' in df.columns and 'View Count' in df.columns:
    top10 = df.nlargest(10, 'View Count')

    plt.figure(figsize=(8,8))
    plt.pie(
        top10['View Count'],
        labels=top10['Channel Name'],
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.viridis(range(0, 256, int(256/len(top10))))
    )

    plt.title('Top 10 YouTube Channels by View Count', fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.DataFrame({
    'Country': ['India', 'USA', 'UK', 'South Korea', 'Brazil',
                'Japan', 'Indonesia', 'Russia', 'Mexico', 'Germany'],
    'View Count': [1100000000, 950000000, 870000000, 920000000, 760000000,
                   880000000, 640000000, 700000000, 600000000, 550000000],
    'Subscribers': [10000000, 8500000, 7800000, 8200000, 7500000,
                    7300000, 7100000, 6900000, 6700000, 6500000]
})


if all(col in df.columns for col in ['Country', 'View Count', 'Subscribers']):
    plt.figure(figsize=(10,6))
    plt.scatter(
        df['Subscribers'],
        df['View Count'],
        s=150,
        color='mediumseagreen',
        edgecolor='black'
    )


    for i, row in df.iterrows():
        plt.text(row['Subscribers'], row['View Count'] + 15000000, row['Country'],
                 ha='center', fontsize=9)

    plt.title('Country-wise YouTube Statistics', fontsize=16)
    plt.xlabel('Subscribers', fontsize=12)
    plt.ylabel('View Count', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
